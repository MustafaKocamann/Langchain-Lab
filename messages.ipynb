## Messages
Mesajlar, LangChain'deki modeller için temel bağlam (context) birimidir. Bir LLM (Büyük Dil Modeli) ile etkileşime girerken konuşmanın durumunu temsil etmek için gereken hem içeriği hem de meta verileri taşıyarak modellerin girdisini ve çıktısını temsil ederler.

Mesajlar şu öğeleri içeren nesnelerdir:

* Rol (Role): Mesaj tipini tanımlar (örneğin; sistem, kullanıcı).

* İçerik (Content): Mesajın asıl içeriğini temsil eder (metin, resimler, ses, belgeler vb. gibi).

* Meta Veri (Metadata): Yanıt bilgileri, mesaj kimlikleri (ID) ve token kullanımı gibi isteğe bağlı alanlardır.

* LangChain, çağrılan modelden bağımsız olarak tutarlı bir davranış sergilenmesini sağlayan ve tüm model sağlayıcılarında çalışan standart bir mesaj tipi sunar.

```python
import os 
from langchain.chat_models import init_chat_model
from langchain_groq import ChatGroq
from dotenv import load_dotenv
load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
)
```

```python
llm.invoke("Yapay zeka nedir?")  # Bu girdi "HumanMessage" olarak geçer.

## Çıktı ise LLM'den gelen bir yanıt olduğu için "AI Message" olarak geçer. 
```

## Text Prompts 
Metin istemleri karakter dizileridir (strings) konuşma geçmişini tutmanızın gerekmediği, basit ve doğrudan içerik oluşturma görevleri için idealdir.

Özetle: Eğer modelle karşılıklı bir sohbet yerine sadece tek bir soru sorup cevap almak istiyorsanız (örneğin; "Bana bir şiir yaz"), "Text Prompt" kullanmak en pratik yoldur.

```python
llm.invoke("What is LangGraph") ## Bu text HumanMessage olarak değerlendirilmektedir.
```

Metin İstemlerini (Text Prompts) Şu Durumlarda Kullanın:
* Tek ve bağımsız bir isteğiniz olduğunda.

* Konuşma geçmişine (sohbet belleğine) ihtiyaç duymadığınızda.

* Minimum kod karmaşıklığı istediğinizde.

## Mesaj İstemleri (Message Prompts)
Alternatif olarak, bir mesaj nesneleri listesi sunarak modele bir mesaj listesi iletebilirsiniz.

Mesaj Türleri (Message types)
Sistem Mesajı (System message): Modele nasıl davranması gerektiğini söyler ve etkileşimler için bağlam (context) sağlar.

* İnsan Mesajı (Human message): Kullanıcı girdisini ve modelle olan etkileşimleri temsil eder.

* Yapay Zeka Mesajı (AI message): Metin içeriği, araç çağrıları (tool calls) ve meta veriler dahil olmak üzere model tarafından oluşturulan yanıtları temsil eder.

Araç Mesajı (Tool message): Araç çağrılarının (tool calls) çıktılarını temsil eder.

Özetle: Bu yapı, basit metin istemlerinin (Text Prompts) aksine çok daha detaylı ve hiyerarşik bir iletişim kurmanızı sağlar. Özellikle modelin bir araç (fonksiyon) kullandığı senaryolarda "Tool message" kritik bir rol oynar.

## Sistem Mesajı (System Message)
Bir SystemMessage, modelin davranışını hazırlayan başlangıç niteliğindeki bir dizi talimatı temsil eder. Konuşmanın üslubunu (tonunu) belirlemek, modelin rolünü tanımlamak ve yanıtlar için temel kurallar oluşturmak için bir sistem mesajı kullanabilirsiniz.

* Sistem mesajları, yapay zekanın "kişiliğini" belirlediğiniz yerdir. Örneğin:

"Sen nazik bir müşteri temsilcisisin."

* Sistem mesajı (System message): Modele nasıl davranması gerektiğini söyler ve etkileşimler için bağlam (context) sağlar.

* İnsan mesajı (Human message): Kullanıcı girdisini ve modelle olan etkileşimleri temsil eder.

* Yapay Zeka mesajı (AI message): Metin içeriği, araç çağrıları (tool calls) ve meta veriler dahil olmak üzere model tarafından oluşturulan yanıtları temsil eder.

* Araç mesajı (Tool message): Araç çağrılarının çıktılarını temsil eder.

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

messages = [
    SystemMessage("You are an expert poetry expert"), ## Talimat gibidir.
    HumanMessage("Write a poem on exams")
]

response = llm.invoke(messages)
response.content
```

```python
system_msg = SystemMessage("You are a helpful coding assistant.")

messages = [
    system_msg,
    HumanMessage("How do I create REST API?")
]

response = llm.invoke(messages)
print(response.content)
```

```python
from langchain.messages import SystemMessage, HumanMessage

system_msg = SystemMessage("""
You are a senior Python Developer with expertise in web frameworks.)
Always provide code examples and explain your reasoning.
Be concise but through şn your explanations.
""")

messages = [
    system_msg,
    HumanMessage("How do i create for loop")
]

response = llm.invoke(messages)
print(response.content)
```

* Rol (Role): Mesaj tipini tanımlar (örneğin; sistem, kullanıcı).

* İçerik (Content): Mesajın asıl içeriğini temsil eder (metin, resimler, ses, belgeler vb. gibi).

* Meta Veri (Metadata): Yanıt bilgileri, mesaj kimlikleri (ID) ve token kullanımı gibi isteğe bağlı alanlardır.

```python
human_msg = HumanMessage(
    content = "Hello",
    name = "Mustafa",
    id = "0912"
)
```

```python
response =llm.invoke([
    human_msg
])

response
```

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

ai_msg = AIMessage("I'd be happy to help you with that question!")

messages = [
    SystemMessage("You are a helpful assistant"),
    HumanMessage("Can you help me?"),
    ai_msg,
    HumanMessage("Great! What's the 2+2")
]

response = llm.invoke(messages)
print(response.content)
```

```python
response.usage_metadata ## bu şekilde metadata bilgilerine de erişebiliriz.
```

```python
from langchain.messages import AIMessage, ToolMessage
from langchain_core.messages.tool import tool_call_chunk

ai_message = AIMessage(
    content = [],
    tool_calls = [{
        "name": "get_weather",
        "args": {"location": "San Francisco"},
        "id": "call_0912"
    }]
)

weather_result = "Sunny"
tool_message = ToolMessage(
    content = weather_result,
    tool_call_id = "call_0912"
)

messages = [
    HumanMessage("What's the weather in San Francisco?"),
    ai_message,
    tool_message,
]

response = llm.invoke(messages)
print(response.content)
```

## Structured Output
Modellerden, yanıtlarını belirli bir şemaya  uygun bir formatta sunmaları istenebilir. Bu, çıktının kolayca ayrıştırılabilmesini (parse edilmesi) ve sonraki işlem adımlarında kullanılabilmesini sağlamak için yararlıdır. LangChain, yapılandırılmış çıktıyı zorunlu kılmak için birden fazla şema türünü ve yöntemini destekler.

### Pydantic
Pydantic modelleri; alan doğrulaması (field validation), açıklamalar ve iç içe geçmiş yapılar ile en zengin özellik setini sunar.


Bu bölüm, yapay zekanın size sadece "düz metin" cevap vermesi yerine, bir bilgisayar programının doğrudan anlayabileceği (JSON gibi) düzenli veriler üretmesini anlatıyor.

Neden İhtiyaç Duyulur?
Normalde bir LLM'e "Bana 3 tane meyve ismi ver" derseniz, size "Tabii ki, elma, armut ve muz." diyebilir. Ancak bir yazılım geliştiriyorsanız, sadece ["elma", "armut", "muz"] listesine ihtiyacınız vardır. Yapılandırılmış çıktı, modelin gereksiz konuşmadan sadece istediğiniz formatta cevap vermesini sağlar.

Pydantic'in Rolü:
Pydantic, Python'da veri tiplerini kontrol etmek için kullanılan en popüler kütüphanedir. LangChain ile kullanıldığında şu avantajları sağlar:

Doğrulama: Modelden gelen verinin bir sayı mı yoksa metin mi olduğunu kontrol eder.

Hata Yönetimi: Eğer model yanlış bir formatta cevap verirse, Pydantic bunu yakalar.

Karmaşık Yapılar: Bir kullanıcı objesinin içinde adres, adresin içinde de posta kodu gibi iç içe geçmiş (nested) verileri tanımlamanıza olanak tanır.
